```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv', header=0)
```

```python
df.info()
```

```python
unnecessary_columns=['Name', 'Ticket', 'Cabin']
```

```python
df=df.drop(unnecessary_columns, axis=1)
```

```python
df.info()
```

```python
dummies=[]
cols=['Pclass', 'Sex', 'Embarked']
for col in cols:
    dummies.append(pd.get_dummies(df[col]))
```

```python
titanic_dummies=pd.concat(dummies, axis=1)
```

```python
df=pd.concat((df, titanic_dummies), axis=1)
```

```python
df.info()
```

```python
df=df.drop(['Pclass', 'Sex', 'Embarked'], axis=1)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Age']=df['Age'].interpolate()
```

```python
df.info()
```

```python
X=df.values
y=df['Survived'].values
```

```python
import numpy as np
```

```python
#Delete Survived Column
X=np.delete(X, 1, axis=1)
```

```python
from sklearn import cross_validation
#Split Train and Test Set

X_train, X_test, y_train, y_test=cross_validation.train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn import ensemble
clf=ensemble.GradientBoostingClassifier(n_estimators=50)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
test=pd.read_csv('test.csv', header=0)
```

```python
#Remove Unnecessary Columns
test=test.drop(unnecessary_columns, axis=1)
```

```python
test.info()
```

```python
#Convert to Categorical
dummies=[]
cols=['Pclass', 'Sex', 'Embarked']
for col in cols:
    dummies.append(pd.get_dummies(test[col]))
test_dummies=pd.concat(dummies, axis=1)
test=pd.concat((test, test_dummies), axis=1)

#Remove after Converstion
test=test.drop(cols, axis=1)
```

```python
test.info()
```

```python
#Deal with missing value
test['Fare']=test['Fare'].interpolate()
test['Age']=test['Age'].interpolate()
```

```python
test.info()
```

```python
X_results=test.values
```

```python
y_results=clf.predict(X_results)
```

```python
output=np.column_stack((X_results[:, 0], y_results))
```

```python
output.info()
```

```python
output
```

```python
output.shape
```

```python
df_results=pd.DataFrame(output.astype('int'), columns=['PassengerID', 'Survived'])
```

```python
df_results.to_csv('titanic_results.csv', index=False)
```
